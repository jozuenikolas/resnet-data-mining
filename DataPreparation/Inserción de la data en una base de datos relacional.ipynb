```python
my_conn = create_engine("mysql+mysqldb://root:narias@localhost/resnetbdd?charset=utf8mb4")
```

```python
#article_affiliation
dfArticlesAffiliations = pd.read_pickle('data/completo/'+directorio+'/recast_article_affiliation.pkl')
dfArticlesAffiliations
```

```python
#author
dfAuthors = pd.read_pickle('data/completo/'+directorio+'/recast_author.pkl')
dfAuthors
```

```python
#article_autor
dfArticlesAuthors = pd.read_pickle('data/completo/'+directorio+'/recast_article_author.pkl')
dfArticlesAuthors
```

<h4 style="text-align:justify;">Paso 3: Insertar dataframes</h4>

```python
#affiliation
dfAffiliations.to_sql(con=my_conn, name='affiliation', if_exists='append', index=True)
```

```python
#article
columns = ['doi', 'title', 'publication_date', 'abstract', 'author_count', 'affiliation_count']
dfArticles[columns].to_sql(con=my_conn, name='article', if_exists='append', index=True)
```

```python
#article_affiliation
dfArticlesAffiliations.to_sql(con=my_conn, name='article_affiliation', if_exists='append', index=False)
```

```python
#author
dfAuthors.to_sql(con=my_conn, name='author', if_exists='append', index=True)
```

```python
#article_autor
dfArticlesAuthors.to_sql(con=my_conn, name='article_author', if_exists='append', index=False)
```

```python
#author_afiliation
dfAuthorsAffiliations.to_sql(con=my_conn, name='author_affiliation', if_exists='append', index=False)
```

```python
#author_keyword
dfAuthorKeywords.to_sql(con=my_conn, name='author_keyword', if_exists='append', index=True)
```

```python
#article_author_keyword
dfArticlesAuthorKeywords.to_sql(con=my_conn, name='article_author_keyword', if_exists='append', index=False)
```
