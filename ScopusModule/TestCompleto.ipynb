<h4 style="text-align:justify;">Importar librerías</h4>

```python
import json
from Client import Client
import pandas as pd
import numpy as np
from Search import Search
from urllib.parse import quote_plus as url_encode
```

<h4 style="text-align:justify;">Cargar el archivo de configuración</h4>

```python
con_file = open("../config.json")
config = json.load(con_file)
con_file.close()
print(type(config), config)
```

<h4 style="text-align:justify;">Configurar el cliente para realizar peticiones</h4>

```python
client = Client(api_key = config['apikey'], inst_token = config['insttoken'])
```

```python
print(client.api_key)
print(client.auth_token)
print(client.inst_token)
```

<h3 style="text-align:center;">Obtener todas las afiliaciones</h3>

<h4 style="text-align:justify;">Definir el URL para búsqueda de afiliaciones</h4>

```python
#testAffiliationSearch = AffiliationSearch(query="AFFIL(Quito)")
searchType = "affiliation"
#query = url_encode("AF-ID(60072054) OR AF-ID(60104598)")
query = url_encode("AF-ID(60108912) OR AF-ID(60105389)")
view = "STANDARD"
field = "dc:identifier,affiliation-name,city,country,document-count,link"
count = "200"
facets = url_encode("affilcountry(include=Ecuador)")
url = "https://api.elsevier.com/content/search/"+searchType+"?query="+query+"&view="+view+"&count="+count+"&facets="+facets+"&field="+field
testAffiliationSearch = Search(url=url)
```

```python
print(testAffiliationSearch.url)
```

<h4 style="text-align:justify;">Ejecutar la petición</h4>

```python
testAffiliationSearch.execute(client,True)
```

<h4 style="text-align:justify;">Mostrar resultados</h4>

```python
#Número de resultados totales
print('tot_num_res:', testAffiliationSearch.tot_num_res)
#Resultados en formato JSON
testAffiliationSearch.results
```

```python
#Resultados en un dataframe
dfAffiliationSearch = pd.DataFrame(testAffiliationSearch.results)
dfAffiliationSearch
```

<h4 style="text-align:justify;">Limpiar resultados</h4>

```python
#Eliminar columnas innecesarias
dfAffiliationSearch.pop("@_fa")
dfAffiliationSearch.pop("prism:url")
#Renombrar columnas
dfAffiliationSearch.rename(columns = {'dc:identifier': 'identifier', 'affiliation-name': 'affiliation_name', 'document-count': 'document_count'}, inplace = True)
#Modificar valores de columnas
dfAffiliationSearch['identifier'] = dfAffiliationSearch['identifier'].apply(lambda x: x.split(":")[1])
dfAffiliationSearch
```

<h3 style="text-align:center;">Obtener todos los Artículos dado el ID de la Afiliación</h3>

<h4 style="text-align:justify;">Definir el URL para búsqueda de artículos</h4>

```python
searchType = "scopus"
query = url_encode("AF-ID(60108912)")
view = "COMPLETE"
#field = "dc:identifier,dc:title,document-count,affilname,affiliation-country,afid,authname,given-name"
field = "dc:identifier,dc:title,author-count,prism:coverDate,dc:description,authkeywords,afid,authid,given-name,surname"
count = "25"
url = "https://api.elsevier.com/content/search/"+searchType+"?query="+query+"&count="+count+"&field="+field
#"&view="+view+
testArticleSearch = Search(url=url)
```

```python
print(testArticleSearch.url)
```

<h4 style="text-align:justify;">Ejecutar la petición</h4>

```python
testArticleSearch.execute(client)
```

<h4 style="text-align:justify;">Mostrar resultados</h4>

```python
dfArticleSearch = pd.DataFrame(testArticleSearch.results)
dfArticleSearch
```

<h4 style="text-align:justify;">Limpiar resultados</h4>

```python
#Crear DataFrame de autores y número de autores
dfAuthors = dfArticleSearch[['author-count', 'author']].copy()
#Eliminar columnas innecesarias
dfArticleSearch.pop("@_fa")
dfArticleSearch.pop("prism:url")
dfArticleSearch.pop("author-count")
dfArticleSearch.pop("author")
#Renombrar columnas
dfArticleSearch.rename(columns = {'dc:identifier': 'identifier', 'dc:title': 'title', 'prism:coverDate': 'publication_date'}, inplace = True)
#Modificar valores de columnas
dfArticleSearch['identifier'] = dfArticleSearch['identifier'].apply(lambda x: x.split(":")[1])
dfArticleSearch
```

```python
dfAuthors['author-count'] = dfAuthors['author-count'].apply(lambda x: int(x['@total']))
dfAuthors
```

```python
full_authors = pd.DataFrame()
for index, row in dfAuthors.iterrows():
    #full_authors = pd.DataFrame(row['author'][:row['author-count']])
    full_authors = pd.concat([full_authors, pd.DataFrame(row['author'][:row['author-count']])], ignore_index=True)
full_authors
```
