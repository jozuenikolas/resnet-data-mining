```python
articuloResultado = testArticleRetrieval.result.copy()
```

```python
articuloResultado
```

```python
if 'item' in articuloResultado:
    del articuloResultado['item']
    
if 'idxterms' in articuloResultado:
    del articuloResultado['idxterms']
    
if 'language' in articuloResultado:
    del articuloResultado['language']
    
if 'subject-areas' in articuloResultado:
    del articuloResultado['subject-areas']
```

```python
articuloResultado
```

```python
retrievalToSearch = {}
retrievalToSearch['@_fa'] = 'true'
if 'prism:url' in articuloResultado['coredata']:
    retrievalToSearch['prism:url'] = articuloResultado['coredata']['prism:url']
else:
    retrievalToSearch['prism:url'] = np.nan
#dc:identifier
if 'dc:identifier' in articuloResultado['coredata']:
    retrievalToSearch['dc:identifier'] = articuloResultado['coredata']['dc:identifier']
else:
    retrievalToSearch['dc:identifier'] = np.nan
#prism:doi
if 'eid' in articuloResultado['coredata']:
    retrievalToSearch['eid'] = articuloResultado['coredata']['eid']
else:
    retrievalToSearch['eid'] = np.nan
#dc:title
if 'dc:title' in articuloResultado['coredata']:
    retrievalToSearch['dc:title'] = articuloResultado['coredata']['dc:title']
else:
    retrievalToSearch['dc:title'] = np.nan
#prism:coverDate
if 'prism:coverDate' in articuloResultado['coredata']:
    retrievalToSearch['prism:coverDate'] = articuloResultado['coredata']['prism:coverDate']
else:
    retrievalToSearch['prism:coverDate'] = np.nan
#dc:description
if 'dc:description' in articuloResultado['coredata']:
    retrievalToSearch['dc:description'] = articuloResultado['coredata']['dc:description']
else:
    retrievalToSearch['dc:description'] = np.nan
#affiliation
if 'affiliation' in articuloResultado:
    if type(articuloResultado['affiliation']) == list:
        retrievalToSearch['affiliation'] = articuloResultado['affiliation']
    else:
        retrievalToSearch['affiliation'] = [articuloResultado['affiliation']]
else:
    retrievalToSearch['affiliation'] = np.nan
#author
if 'authors' in articuloResultado:
    retrievalToSearch['author'] = articuloResultado['authors']['author']
else:
    retrievalToSearch['author'] = np.nan
#authkeywords
if 'authkeywords' in articuloResultado:
    if articuloResultado['authkeywords'] is not None:
        retrievalToSearch['authkeywords'] = articuloResultado['authkeywords']['author-keyword']
    else:
        retrievalToSearch['authkeywords'] = np.nan
else:
    retrievalToSearch['authkeywords'] = np.nan
```

```python
retrievalToSearch
```

```python
newAuthorKeywords = ''
if type(retrievalToSearch['authkeywords']) != float:
    for item in retrievalToSearch['authkeywords']:
        if '$' in item:
            newAuthorKeywords = newAuthorKeywords + item['$'] + ' | '
    retrievalToSearch['authkeywords'] = newAuthorKeywords[:-3]
```

```python
retrievalToSearch['affiliation']
```

```python
newAffiliation = []
for item in retrievalToSearch['affiliation']:
    newAffiliation.append({
        '@_fa': 'true',
        'affiliation-url': item['@href'],
        'afid': item['@id'],
        'affiliation-country': item['affiliation-country'],
    })
retrievalToSearch['affiliation'] = newAffiliation
```

```python
newAuthor = []
for item in retrievalToSearch['author']:
    newAuthorDict = {}
    newAuthorDict['@_fa'] = 'true'
    newAuthorDict['author-url'] = item['author-url']
    newAuthorDict['authid'] = item['@auid']
    if 'affiliation' in item:
        if type(item['affiliation']) == list:
            newAuthorDict['afid'] = [{'@_fa':'true','$':afil['@id']} for afil in item['affiliation']]
        else:
            newAuthorDict['afid'] = [{'@_fa':'true','$':item['affiliation']['@id']} ]
    else:
        newAuthorDict['afid'] = []
    newAuthor.append(newAuthorDict)
    
retrievalToSearch['author'] = newAuthor
```
