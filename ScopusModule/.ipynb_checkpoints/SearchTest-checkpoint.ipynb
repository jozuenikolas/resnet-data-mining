```python
def chunker(seq, size):
    return (seq[pos:pos + size] for pos in range(0, len(seq), size))
```

```python
animals = ['cat', 'dog', 'rabbit', 'duck', 'bird', 'cow', 'gnu', 'fish']

for group in chunker(authorsId, 10):
    print(len(group))
```

```python
dataAuthorSearch = []
view = "STANDARD"
field = "dc:identifier,eid,orcid,surname,given-name,affiliation-id,document-count"
for group in chunker(authorsId,50):
    searchType = "author"
    query = url_encode(' OR '.join(["AU-ID("+item+")" for item in group]))
    url = "https://api.elsevier.com/content/search/"+searchType+"?query="+query+"&field="+field
    authorSearch = Search(url=url)
    print(authorSearch.url)
    authorSearch.execute(client, True)
    print('tot_num_res:', authorSearch.tot_num_res)
    dataAuthorSearch = dataAuthorSearch + authorSearch.results
```
