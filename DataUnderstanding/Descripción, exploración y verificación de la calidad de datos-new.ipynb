<h3><strong>Parte 2: Recast de las tablas</strong></h3>

```python
#Modifica el valor de la columna affiliation.
dfArticles['affiliation'] = dfArticles['affiliation'].apply(lambda x: Utils.rewriteArticleAffilList(x) 
                                                            if type(x) == list else np.nan)
#Modifica el valor de la columna author. 
dfArticles['author'] = dfArticles['author'].apply(lambda x: Utils.rewriteArticleAuthors(x) 
                                                    if type(x) == list else np.nan)
```

```python


#Agrega la columna author_count que contiene el número de autores de un artículo.
dfArticles['author_count'] = dfArticles['author'].apply(lambda x: int(len(x))
                                                       if type(x) == list else 0)

#Agrega la columna affiliation_count que contiene el número de afiliaciones de un artículo.
dfArticles['affiliation_count'] = dfArticles['affiliation'].apply(lambda x: int(len(x))
                                                                 if type(x) == list else 0)
```

```python
idsDiscordiaArticulos = []
for index, row in dfArticles.iterrows():
    authorAfid = []
    affilAfid = []
    if type(row['author']) == list: 
        for author in row['author']:
            for afid in author['afid']:
                if afid not in authorAfid: authorAfid.append(afid)
    if type(row['affiliation']) == list: 
        for affiliation in row['affiliation']:
            affilAfid.append(affiliation['afid'])
    if set(affilAfid) != set(authorAfid):
        idsDiscordiaArticulos.append(row['dc:identifier'])
        if len(authorAfid) < len(affilAfid):
            print('ID:',row['dc:identifier'])
            print("Título:",row['dc:title'])
            print("Número de autores",row["author_count"])
            print("Número de afiliaciones de autores del artículo", len(authorAfid))
            print("Número de afiliaciones del artículo", len(affilAfid))
            print("Afiliaciones diferentes")
            print(np.setdiff1d(authorAfid,affilAfid))
            print(np.setdiff1d(affilAfid,authorAfid), '\n')
        
```
