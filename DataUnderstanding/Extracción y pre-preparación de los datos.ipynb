<h3><strong>Parte 2: Extraer los artículos de Ecuador</strong></h3>

<h4 style="text-align:justify;">Paso 1: Definir el URL para la búsqueda de artículos</h4>

```python
#Tipo de búsqueda
searchType = "scopus"
#Lista de elementos que se devolverán en la respuesta
view = "COMPLETE"
#Nombre de campos específicos que deben retornarse
field = "dc:identifier,doi,dc:title,coverDate,dc:description,authkeywords,afid,affilname,affiliation-city,affiliation-country,authid,authname,given-name,surname,initials"
#Número máximo de resultados que se devolverán por petición
count = "25"
#Parámetro se utiliza para ejecutar una búsqueda de paginación profunda
cursor = '*'
#Búsqueda booleana que se ejecutará en el clúster de SCOPUS.
query = url_encode("AFFIL(AFFILCOUNTRY(Ecuador))")
#Url de búsqueda de artículos
url = "https://api.elsevier.com/content/search/"+searchType+"?query="+query+"&count="+count+"&view="+view+"&field="+field+"&cursor="+cursor
print("url:", url)
```

<h4 style="text-align:justify;">Paso 2: Declarar el objeto para la búsqueda de artículos a través de la API abstract search</h4>

```python
articleSearch = Search(url=url)
```

<h4 style="text-align:justify;">Paso 3: Ejecutar la petición</h4>

```python
inicioArticle = time.time()
articleSearch.execute(client, True)
finArticle = time.time()
```

<h4 style="text-align:justify;">Paso 4: Mostrar la fecha y tiempo de ejecución de la petición</h4>

```python
print('Fecha de inicio:', time.ctime(inicioArticle)) 
print('Fecha de fin:', time.ctime(finArticle)) 
print('Tiempo total de ejecución:', finArticle-inicioArticle, 'segundos.') 
```

<h4 style="text-align:justify;">Paso 3: Cargar los datos de los artículos en un dataframe</h4>

```python
#Resultados en un dataframe
dfArticles = pd.DataFrame(articleSearch.results)
dfArticles
```

<h4 style="text-align:justify;">Paso 7: Obtener los IDs de los artículos con autores incompletos</h4>

```python
articlesIdsIncompleteAuth = []
for index, row in dfArticles.iterrows():
    if type(row['author']) == list:
         if len(row['author']) > 98: 
                articlesIdsIncompleteAuth.append(row['dc:identifier'].split(':')[1])
```

```python
print('Numero de artículos con autores incompletos:', len(articlesIdsIncompleteAuth))
```

<h4 style="text-align:justify;">Paso 8: Extraer los artículos dado su ID a través de la API abstract retrieval</h4>

```python
inicioArticleRetrieval = time.time()
dataArticles = []
#Representa la lista de elementos que se devolverán en la respuesta
view = "FULL"
for articleId in articlesIdsIncompleteAuth:
    #URL para la recuperación de artículos
    url = "https://api.elsevier.com/content/abstract/scopus_id/"+articleId+"?view="+view
    #Declaración del objeto para la recuperación de artículos a través de la API abstract retrieval
    articleRetrieval = ArticleRetrieval(url=url)
    #Ejecución de la(s) petición(es)
    articleRetrieval.execute(client)
    #Concatenación de los resultados
    dataArticles.append(
        Utils.articleRetrievalToScopusSearch(articleRetrieval.result)
    )
    print('Artículos recuperados de momento:', len(dataArticles))
finArticleRetrieval = time.time()
```

<h4 style="text-align:justify;">Paso 9: Mostrar la fecha y tiempo de ejecución de las peticiones</h4>

```python
print('Fecha de inicio:', time.ctime(inicioArticleRetrieval)) 
print('Fecha de fin:', time.ctime(finArticleRetrieval)) 
print('Tiempo total de ejecución:', finArticleRetrieval-inicioArticleRetrieval, 'segundos.') 
```

<h3><strong>Parte 3: Extraer las afiliaciones ecuatorianas de los artículos</strong></h3>

<h4 style="text-align:justify;">Paso 1: Obtener los datos de las afiliaciones </h4>

<h4 style="text-align:justify;">Paso 2 Eliminar registros duplicados</h4>
